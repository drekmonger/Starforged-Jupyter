```python
import json, re
data = dict()
```

```python
def format_move_text (text):
    first_letter = text[0].capitalize
    text = text.strip()
    #text = text.replace("**", "*")
    text = text.replace("__", "*")
    #text = text.replace("â€™", "'")
    #text = text.replace("â€˜", "'")
    
    #remove tables
   
    #table_index = text.find("Roll  ")
    #if  not table_index == -1:
    #    text = text[:table_index]
    
    return text[0].capitalize() + text[1:]
```

```python
moves = {}

#moves["Categories"] = []
      
with open("starforged_moves.json", "r") as file:
    move_data = json.load(file)

for category in move_data.keys():
    cat_name = re.findall("(.+?) Moves", category)
    
    if len(cat_name) == 1:
        print ("----" + cat_name[0])
        #moves["Categories"].append(cat_name[0])
      
        for move in move_data[category].keys():
            
            print (move.title())
        
            text = move_data[category][move]["Text"]
            
            look_for_hits = re.findall(r'(.+?)\n\nOn a \*\*strong', text, re.DOTALL)
            if len(look_for_hits) == 0:    
                strong = None
                weak = None
                miss = None
                kind = "No Roll"
            else:
                strong = format_move_text(re.findall(r'On a \*\*strong hit\*\*, (.+?)On a', text, re.DOTALL)[0])
                weak = format_move_text(re.findall(r'On a \*\*weak hit\*\*, (.+?)On a', text, re.DOTALL)[0])
                miss = format_move_text(re.findall(r'On a \*\*miss\*\*, (.+)', text, re.DOTALL)[0])
                if "***Progress Move***" in text:
                    kind = "Progress"
                else:
                    kind = "Action"
                    
            
            moves[move.title()] = {"Name": move.title(), "Category": cat_name[0], "Kind": kind, "Text": format_move_text(text), 
                           "Strong": strong, "Weak": weak, "Miss": miss }
    
with open("starforged_moves_drek.json", "w") as file:
    json.dump(moves, file)
    
```

```python
cat_list = ["campaign_start", "character", "creature", "interpretive", "location", "moves", "planet", "settlement", "starship"]
```

```python
oracles = {}

world_table = []

for cat in cat_list:
    with open(f"starforged_oracles_{cat}.json", "r", encoding="utf-8") as file:
        cat_data = json.load(file)
    
    print ("--" + cat)    
        
    
    for key, item in cat_data["Oracles"].items():
        
        
        
            
        print (key)
            
        subtables = type(item) == dict
        
        region_required = False
        world = False
        if subtables:
            print ("**subtables")
            if "World" in key:
                world = True
                print ("***world")
                world_table.append({"Chance": item["Chance"], "Description": key})
            
            if "Outlands" in item.keys():
                region_required = True
                print ("***region required")
        
        
        if key == "First Look":
            key = cat.title() + " First Look"
            
        oracles[key] = {"table": item,  "subtables": subtables, "region required": region_required, "world": world, "category": cat}

        
oracles["Planet Type"] = {"table": world_table, "subtables": False, "region required": False, "world": False, "category": "planet"}

with open(f"starforged_oracles_drek.json", "w", encoding="utf-8") as file:
    json.dump(oracles, file) 
```
